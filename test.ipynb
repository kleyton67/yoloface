```python
import os
import torch
import numpy as np
from PIL import Image
from face_detector import YoloDetector

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
os.environ["CUDA_VISIBLE_DEVICES"]=""
model = YoloDetector(target_size=240, device='cpu', min_face=40)
orgimg = np.array(Image.open('photo_2022-12-13_08-38-58.jpg'))
bboxes, points, score = model.predict(orgimg)
print(score)
```

```python
image_crop, Affine_M = model.align(orgimg, points)[0]
```

```python
%matplotlib inline 
from matplotlib import pyplot as plt
plt.imshow(image_crop, interpolation='nearest')
plt.show()
```
